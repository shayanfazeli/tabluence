# Label Distribution

With a specific stress criteria (in this case: `x['overall_quantized_stress_value'] > 0.5`),  this notebook demonstrates how we can observe the label distribution:

```python
import matplotlib.pyplot as plt
import numpy
import torch
import pickle
import gzip
import pandas
```

Loading the data manager cache:

```python
with gzip.open('dataset_cache/dataset-cache-1.pkl.gz', 'rb') as handle:
    cache = pickle.load(handle)
```

```python
cache[0][0]
```

```python
metadata_df = pandas.DataFrame(cache[0])
```

```python
metadata_df.head()
```

```python
metadata_df['label'] = metadata_df.apply(lambda x: x['overall_quantized_stress_value'] > 0.5, axis=1)
```

```python
metadata_df.groupby(['subject_id', 'label']).count().loc[:, ['overall_stress_value']]
```
